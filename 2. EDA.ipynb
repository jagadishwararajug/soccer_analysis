```python
import pandas as pd
import matplotlib as plt
import numpy as np
import sweetviz as sv
pd.set_option('display.max_rows', 50)
pd.set_option('display.max_columns', 500)
```

```python
df = pd.read_csv('preproc.csv')
df
```

```python
extras = ['ID', 'Name',  'Nationality', 'Club']

y = 'Overall'

skills = [ 'Crossing', 'Finishing', 'HeadingAccuracy', 
       'ShortPassing', 'Volleys', 'Dribbling',
       'Curve', 'FKAccuracy', 'LongPassing', 'BallControl', 'Acceleration',
       'SprintSpeed', 'Agility', 'Balance', 'ShotPower',
       'Jumping', 'Stamina', 'Strength', 'LongShots', 'Aggression',
       'Interceptions', 'Positioning', 'Vision', 'Penalties', 'Composure',
       'Marking', 'StandingTackle', 'SlidingTackle', 'GKDiving', 'GKHandling',
       'GKKicking', 'GKPositioning', 'GKReflexes']

cols = [ 'Defense', 'Forward', 'Goalkeeper', 'Midfield', 'Age', 'Height', 'Weight' ] + skills + [y]
```

```python
df = df[cols].astype('float32')
df
```

### Exploring the distribution of each column

```python
analyze_report = sv.analyze(df)
analyze_report.show_html('analyze.html', open_browser=False)
```

#### Observations

**Skills**
* Most skill related attributes have peaks between 60-70. This is where most players skill levels will lie.
* Most skill related attributes drop significantly after 80. Hence, any player with a skill value above 80 for a skill is highly rated in that skill.
* Skill values below 40 are rare in most skills. Hence, a value below 40 indicates that a player is not very good at a skill.

**Overall Ratings**
* Most players have a rating between 60 to 70. A rating above 80 is a high rating. And a rating below 55 is a low rating. 

### Correlation matrix

```python
corr = df.corr()
corr.style.background_gradient(cmap='coolwarm')
```

```python
# removing repeated correlations Ex: (A, B) and (B, A)
filter_corr = np.triu(np.ones(corr.shape)).astype('bool').reshape(corr.size)
corr = corr.stack()[filter_corr]
sorted_corr = corr.sort_values()
# removing values of correlation of a column with itself 
sorted_corr = sorted_corr[[e for e in sorted_corr.index if e[0] != e[1]]]
sorted_corr
sorted_corr
```

#### Top 50 positive correlations

```python
sorted_corr.iloc[-50:]
```

### Observations

**Goalkeeper related correlations**
* The position **Goalkeeper is highly positively correlated with Goalkeeper related skills**. This is to be expected since goalkeepers need goalkeeper related skills.

**Defense and Midfield related correlations**
* **StandingTackle and SlidingTackle** are highly correlated. These are skills that are useful in Defense and Defenders need to be good at both to defend well.
* **Interception and StandingTackle** are also defense related skills and is something a defender and midfielders can be expected to be good at.
* **Marking and StandingTackle** is a combination that midfielders often use to steal the ball from the opposition.

**Forward related correlations**
* **Dribbling and BallControl** is useful for both Forward and Midfielders.
* **Finishing and Positioning** is essential to score goals.
